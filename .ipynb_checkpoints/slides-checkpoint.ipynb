## Dataset Introduction and Acquisition
- Data concern stats for cards in the deckbuilder mobile game *Clash Royale*
- Cards' recent popularity and match success rates are catalogued on the official Clash Royale API
- Usage data were scraped along information on cards' stats and other characteristics

## Data Cleaning and EDA
- Some null values observed in evolution level, damage, and health
    - Cards without an evolution level were imported as NaN values and had to be manually recoded
    - Some card types were imported as NaN for health because they lacked any
    - Most missing damage values were corrected manually using game knowledge
- Graphing initial values
    - Cards with high damage rarely also have high health
    - Most cards are troop-type, with building and spell cards being rarer
    - The most commonly-used cards have a middling win rate
- Several dummy variables created
    - Rarity could be mapped to a 0–5 scale
    - Card type, targeting type, movement type, etc. were converted to boolean dummy variables

## Clustering Cards
- Choosing parameters
    - K-prototype was used instead of K-means because data contain mixed categorical and numeric columns
    - 4 was chosen as the number of clusters for troop cards based on the elbow method
    - Buildings and spells were given their own clusters separate from the model (6 clusters total)
    - Health, damage, and cost were the examined variables
- Graphing the clusters
    - 3D visualizations of damage, hit points, and elixir cost created with the `plotly.express` library
- Labeling clusters
    - 0: glass cannons (high-damage, low-health, medium-cost), use to deal damage
    - 1: mini-tanks (low-damage, mid-health, medium-cost), used as cheaper costs
    - 2: tanks (low-damage, high-health, high-cost), used to attack towers
    - 3: cycle cards (low-damage, low-health, low-cost), used as fodder to access more damaging cards
    - 4: buildings, most closely resembling tanks
    - 5: spells, similar damage and cost to cycle cards but 0 health
