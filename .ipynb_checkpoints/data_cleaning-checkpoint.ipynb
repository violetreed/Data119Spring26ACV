```python
import pandas as pd
import numpy as np
```

# Card Data

```python
# Import merged df
cards_df = pd.read_csv('./data/cards_merged.csv')
cards_df
```

```python
# Count nulls
cards_clean_df = cards_df.copy()
cards_df.isnull().sum()
```

## Investigate `maxEvolutionLevel`

```python
# maxEvolutionLevel values
cards_df[['name', 'maxEvolutionLevel']].head(10)
```

| maxEvolutionLevel | Trait |
| ----------------- | ----- |
| NaN | No trait |
| 1.0 | Evolution |
| 2.0 | Hero |
|3.0 | Evolution & Hero |

```python
# Transform maxEvolutionLevel to categorical
cards_clean_df.rename(columns={'maxEvolutionLevel': 'wild'}, inplace=True)
cards_clean_df['wild'] = cards_df['maxEvolutionLevel'].replace({np.nan: 'normal', 1: 'evo', 2: 'hero', 3: 'both'})
cards_clean_df.head(10)
```

```python
cards_clean_df['wild'].isnull().sum()
```
